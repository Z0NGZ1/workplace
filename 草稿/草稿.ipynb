```python
{ 参数设置}
N := 20;      { 波动率周期}
VN := 20;     { 成交量均线周期}
THRESH := 2;  { 均线波动阈值}

{ 价格波动指标}
MA5 := MA(C,5);
OC_L := IF(C > O, O, C);
PRICE_VOLATILITY := AVEDEV(OC_L, N) / MA(OC_L, N) * 100;
MA5_VOLATILITY := AVEDEV(MA5, N) / MA(MA5, N) * 100;

{ 标准差筛选}
STD_RATIO := STD(C, N) / MA(C, N);
STD_RANK := LLV(STD_RATIO, 960) / HHV(STD_RATIO, 960);

{ 成交量筛选}
MA_V := MA(V, VN);
VOL_RANK := MA_V / HHV(MA_V, 1000);

{ 复合条件}
BASIC_COND := STD_RANK < 0.16 AND VOL_RANK < 0.16;

{ 趋势过滤条件}
TREND_FILTER := 
    MA(C,5) > MA(C,10) AND 
    MA(C,10) > MA(C,30) AND 
    C > MA(C,30);

{ 突破量能确认}
VOLUME_CONFIRM := V > MA(V,5) AND V > REF(V,1)*1.2;

{ 最终信号}
BUY_SIGNAL := BASIC_COND AND 
             MA5_VOLATILITY < THRESH AND 
             TREND_FILTER AND 
             VOLUME_CONFIRM;

{ 图形标记}
PLOT(BUY_SIGNAL, "底部信号", COLORGREEN, STYLE_HISTOGRAM);
DRAWICON(BUY_SIGNAL, L-0.5, ICON_BUY);
```
